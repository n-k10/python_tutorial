### 2-3 長方形を動かす

作成したパドルをキーボード操作で動くようにしましょう！

単純な指針として次のような流れとしましょう．

1. ユーザーがキーボードで方向キーを入力
2. その入力を受けたら長方形の座標を少し移動させる
3. 移動させた長方形の座標を `draw` 関数で描画

### Task 2-3-1

上記の指示に従い，関数 `move`，`left`，`right` を記述しなさい．ただし，先に `draw` 関数の拡張を行う必要がある．

### drawの拡張要件

- 関数 `draw` の外側に変数として，現在の座標を表す `now_x`，`now_y` をもつ (関数の中で変数をいじった結果を全体でも使用したいため)
- `now_x`の初期値は275，`now_y`の初期値は450

### moveの要件

- 引数・返り値無し
- 矢印キーの右が押されたら 関数 `right`を呼び出す (`bind`関数を調べて用いる)
- 矢印キーの左が押されたら 関数 `left`を呼び出す (`bind`関数を用いる)
- 関数を作成したら適切なところで呼び出す

### leftの要件

- 引数に `event` という全く使わない変数を入れる (`bind`関数で呼び出す際に参照するため)
- 返り値無し
- 外部にある `now_x` を用いるために`global now_x`と記述
- `now_x` を -2 する
- 上記の処理が終わったら `draw` 関数を呼び出す．

### rightの要件
- 引数に `event` という全く使わない変数を入れる (`bind`関数で呼び出す際に参照するため)
- 返り値無し
- 外部にある `now_x` を用いるために`global now_x`と記述
- `now_x` を +2 する
- 上記の処理が終わったら `draw` 関数を呼び出す．

<details>
<summary>答えのコード</summary>
<code>
import tkinter as tk


now_x = 275
now_y = 450

root = tk.Tk()
root.title("球当てゲーム")
root.geometry("600x480")
cv = tk.Canvas(root, width=600, height=480)
cv.pack()

def draw():
    cv.create_rectangle(now_x, now_y, now_x + 50, now_y + 8, fill='blue', tags="paddle")

def right(event):
    global now_x
    now_x += 2
    draw()

def left(event):
    global now_x
    now_x -= 2
    draw()

def move():
    root.bind("<Right>", right)
    root.bind("<Left>", left)

draw()

def gameloop():
    move()
    root.after(50, gameloop)

gameloop()
root.mainloop()
</code>

### 3-1 クラスによる記述のメリット

ここから先，ボールやブロックの実装を行う必要がありますが，それらも描画処理を記述することになります．しかし，同じコード内で同じ名前を使用することはできません．なので `draw_paddle` や `draw_ball` と記述する必要があります．

また，ひたすら関数で記述すると変数の扱いが大変です．例えば，ブロックの場合はそれぞれが座標を持つことになるため，ブロック一つ一つの座標をリストなどで管理する必要があります．イメージがつかない場合は先ほどの `now_x` と `now_y` を10個以上記述しないといけない，と考えると良いです． ( 見にくいし書いてて辛い )

これらの問題点を解決する書き方として **クラス** というものがあります．これは，それぞれのオブジェクトにプログラムの関数や変数を持たせる，という書き方です．例えば，ブロックは「位置情報，描画処理，削除処理」を記述することになりますが，クラスの考えを使うと，ブロック一つ一つが「位置情報，描画処理，削除処理」を持つことができます．しかもコードは数行で済みます．
